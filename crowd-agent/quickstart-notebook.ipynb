## Run local inference on an image
Replace 'sample.jpg' with your test frame.

```python
import json, subprocess, sys, os
from pathlib import Path
img = 'sample.jpg'  # put a sample image next to the notebook
proc = subprocess.run([sys.executable, 'local/inference_local.py', img], capture_output=True, text=True)
print(proc.stdout)
cv = json.loads(proc.stdout) if proc.returncode == 0 else {'people_count': 0}
```

## Get a decision using the Bedrock stub
This simulates the LLM decision logic offline.

```python
import json, subprocess, sys
p = subprocess.run([sys.executable, 'local/bedrock_stub.py'], input=json.dumps(cv), text=True, capture_output=True)
print(p.stdout)
```
